```python
import youtube_dl
import pandas as pd
import numpy as np
```

# Collecting data

```python
queries = ['machine learning', 'data science', 'kaggle']
```

```python
resultados = []

for query in queries:
    with youtube_dl.YoutubeDL({"ignoreerrors": True}) as ydl:
        vds = ydl.extract_info("ytsearchdateall:{}".format(query), download=False)
        for entry in vds['entries']:
            if entry is not None:
                entry['query'] = query
        resultados += vds['entries']
```

```python
valid_results = [result for result in resultados if result is not None]
df = pd.DataFrame(valid_results)

df['upload_date'] = pd.to_datetime(df['upload_date'])
df['tempo_desde_pub'] = (pd.to_datetime('2020-09-09') - df['upload_date']) / np.timedelta64(1, 'D')
```
